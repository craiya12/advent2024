```python
import os
import pandas as pd
import numpy as np
import re



rules = []
pages = []

with open("advent_5.txt", "r") as f:
    # Read the entire file content and split on a double newline to separate rules from pages
    sections = f.read().strip().split("\n\n")

# The first section is the rules block, the second is the pages block
rules_block, pages_block = sections

# Process the rules, splitting each line by '|'
for line in rules_block.split("\n"):
    # Convert each part to int and append to rules list
    rules.append([int(x) for x in line.split("|")])

# Process the pages, splitting each line by ','
for line in pages_block.split("\n"):
    # Convert each part to int and append to pages list
    pages.append([int(x) for x in line.split(",")])

```

```python
good=[]

rules_d={}


for item in rules:

    if item[0] in rules_d:
        rules_d[item[0]].append(item[1])
        
    else:
        rules_d[item[0]]=[item[1]]

        
#print(rules_d)

```

```python
valid_pages=[]

#pages=[[61, 13, 29]]

for line in pages:
    
        
    progress_list=[]

    
    
    #print(applied_rule_set)


    for num in line:
        #print("check",num)      

        if num in rules_d:
            num_set=rules_d[num]
            #print(num_set)

            #print(num)
            #print("intersection:",set(progress_list).intersection(num_set))


            if set(progress_list).intersection(num_set):
                break

        


        progress_list.append(num)
    
    else:


        valid_pages.append(line)
            

        
#print(valid_pages)
    
middle_pages=[]
    
for line in valid_pages:
    length=len(line)
    
    middle=length//2
    
    middle_pages.append(line[middle])
    
print(sum(middle_pages))
    
```
