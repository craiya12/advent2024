```python

file=open('advent_1_location_id.txt','r')

content=file.readlines()

print(content)
```

```python
location1=[]
location2=[]

d=[]

for line in content:
    data=line.split()
    

    location1.append(int(data[0]))
    location2.append(int(data[1]))
    
loc1_sort=sorted(location1)
loc2_sort=sorted(location2)

for i in range(len(loc1_sort)):
    dif=abs(loc1_sort[i]-loc2_sort[i])
    d.append(dif)
    
final=sum(d)

print(final)
    
```
